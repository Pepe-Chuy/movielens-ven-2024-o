```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_ratings = pd.read_csv("data/ml-latest-small/ratings.csv")
df_ratings.head()
```

```python
df_ratings.describe()
```

- We have 100836 different registries
- 610 users
- the avg rating is 3.5

```python
rating_counts = df_ratings['rating'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings')
plt.show()
```

- the most relevant groups are 3, 3.5, 4 and 5 stars193609

```python
plt.figure()
sns.barplot(data=df_ratings, x='rating', y='movieId')
plt.show()
```
